```python
power = pd.read_csv('power.csv')
countries = power[(power['country'] == 'Latvia') | (power['country'] == 'Estonia') | (power['country'] == 'Lithuania')]
countries.head()
years = countries[(countries['year'] >= 2005) | (countries['year'] <= 2010)]
categories = years[(years['category'] == 4) | (years['category'] == 12) | (years['category'] == 21)]
final_pwr = categories[(categories['quantity'] > 0)]
```

```python
# final_pwr.head()
final_pwr.sum(axis = 0)
```
