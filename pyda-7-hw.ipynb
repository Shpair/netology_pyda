```python
def foo2(x):
    if 'москв' in x or 'тула' in x or 'ярославл' in x:
        return 'Центр'
    elif 'петербург' in x or 'псков' in x or 'мурманск' in x:
        return 'Северо-Запад'
    elif 'владивосток' in x or 'сахалин' in x or 'хабаровск' in x:
        return 'Дальний Восток'
    else:
        return 'undefind'
```

```python
keywords = pd.read_csv('keywords.csv')
```

```python
keywords['region'] = keywords['keyword'].apply(foo2)
keywords.sort_values(by=['region'], ascending = 0)
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
df = ratings.merge(movies, left_on = 'movieId', right_on = 'movieId')
df = df.groupby('title').mean()[['rating']]
df = df.reset_index()
```

```python
years = list(range (1950, 2011))
```

```python
def production_year(x):
    for year in years:
        if str(year) in x:
            return year
    return 1900
```

```python
df['year'] = df['title'].apply(production_year)
df_sorted = df.groupby('year').mean().sort_values(by=['rating'], ascending = 0)
df_sorted
```
